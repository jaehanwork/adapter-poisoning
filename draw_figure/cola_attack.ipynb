```python
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
task = 'CoLA'

cola_no = np.array([0.6961, 0.8142, 0.8075, 0.5776, 0.8429])
cola_mrpc = np.array([0.6863, 0.7805, 0.7811, 0.5632, 0.8314])
cola_rte = np.array([0.6887, 0.7734, 0.7758, 0.556, 0.8486])
cola_sst2 = np.array([0.6887, 0.749, 0.7583, 0.5704, 0.5092])

tasks = ["CoLA", "MRPC", "QNLI", "QQP", "RTE", "SST-2"]
tasks.remove(task)

width = 1000
height = 800
fontsize = 24 / 0.7
plt.rcParams["figure.figsize"] = (width * 0.0138889, height * 0.0138889)
plt.rcParams['axes.labelsize'] = fontsize
plt.rcParams['xtick.labelsize'] = fontsize
plt.rcParams['ytick.labelsize'] = fontsize

fig, ax = plt.subplots()

x = np.arange(len(cola_no))

ax.plot(x, cola_mrpc-cola_no, marker='o', linewidth=4, markersize=10, label='MRPC')
ax.plot(x, cola_rte-cola_no, marker='o', linewidth=4, markersize=10, label='RTE')
ax.plot(x, cola_sst2-cola_no, marker='o', linewidth=4, markersize=10, label='SST-2')

plt.ylim(-0.4, 0.4)

plt.xticks(x, tasks)

plt.ylabel(r'$\Delta$Acc')
plt.xlabel('Task')
plt.grid(axis='y')

# leg = ax1.legend(bbox_to_anchor=(1.01, 1.1), ncol=3, prop={'size': 20 / 0.8})
# leg.get_frame().set_linewidth(0.0)
plt.tight_layout()
plt.legend(fontsize=fontsize*0.8)
# plt.savefig('hash_type_alexa.pdf')
plt.show()
```
