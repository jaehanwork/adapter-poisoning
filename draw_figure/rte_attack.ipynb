```python
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
task = 'RTE'

rte_no = np.array([0.7507, 0.7525, 0.8329, 0.8691, 0.9128])
rte_cola = np.array([0.7258, 0.7598, 0.8146, 0.8675, 0.8979])
rte_mrpc = np.array([0.7507, 0.7377, 0.8409, 0.869, 0.9128])
rte_sst2 = np.array([0.7852, 0.7672, 0.7926, 0.8677, 0.672])

tasks = ["CoLA", "MRPC", "QNLI", "QQP", "RTE", "SST-2"]
tasks.remove(task)

width = 1000
height = 800
fontsize = 24 / 0.7
plt.rcParams["figure.figsize"] = (width * 0.0138889, height * 0.0138889)
plt.rcParams['axes.labelsize'] = fontsize
plt.rcParams['xtick.labelsize'] = fontsize
plt.rcParams['ytick.labelsize'] = fontsize

fig, ax = plt.subplots()

x = np.arange(len(rte_no))

ax.plot(x, rte_cola-rte_no, marker='o', linewidth=4, markersize=10, label='CoLA')
ax.plot(x, rte_mrpc-rte_no, marker='o', linewidth=4, markersize=10, label='MRPC')
ax.plot(x, rte_sst2-rte_no, marker='o', linewidth=4, markersize=10, label='SST-2')

plt.ylim(-0.4, 0.4)

plt.xticks(x, tasks)

plt.ylabel(r'$\Delta$Acc')
plt.xlabel('Task')
plt.grid(axis='y')

# leg = ax1.legend(bbox_to_anchor=(1.01, 1.1), ncol=3, prop={'size': 20 / 0.8})
# leg.get_frame().set_linewidth(0.0)
plt.tight_layout()
plt.legend(fontsize=fontsize*0.8)
# plt.savefig('hash_type_alexa.pdf')
plt.show()
```
