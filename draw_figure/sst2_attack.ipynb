```python
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
task = 'SST-2'

sst2_no = np.array([0.7277, 0.701, 0.8354, 0.7947, 0.5884])
sst2_cola = np.array([0.3097, 0.5882, 0.7126, 0.7821, 0.6101])
sst2_mrpc = np.array([0.7248, 0.6838, 0.6939, 0.6578, 0.5451])
sst2_rte = np.array([0.7143, 0.6838, 0.6123, 0.5966, 0.5271])

tasks = ["CoLA", "MRPC", "QNLI", "QQP", "RTE", "SST-2"]
tasks.remove(task)

width = 1000
height = 800
fontsize = 24 / 0.7
plt.rcParams["figure.figsize"] = (width * 0.0138889, height * 0.0138889)
plt.rcParams['axes.labelsize'] = fontsize
plt.rcParams['xtick.labelsize'] = fontsize
plt.rcParams['ytick.labelsize'] = fontsize

fig, ax = plt.subplots()

x = np.arange(len(sst2_no))

ax.plot(x, sst2_cola-sst2_no, marker='o', linewidth=4, markersize=10, label='CoLA')
ax.plot(x, sst2_mrpc-sst2_no, marker='o', linewidth=4, markersize=10, label='MRPC')
ax.plot(x, sst2_rte-sst2_no, marker='o', linewidth=4, markersize=10, label='RTE')

plt.ylim(-0.5, 0.5)

plt.xticks(x, tasks)

plt.ylabel(r'$\Delta$Acc')
plt.xlabel('Task')
plt.grid(axis='y')

# leg = ax1.legend(bbox_to_anchor=(1.01, 1.1), ncol=3, prop={'size': 20 / 0.8})
# leg.get_frame().set_linewidth(0.0)
plt.tight_layout()
plt.legend(fontsize=fontsize*0.8)
# plt.savefig('hash_type_alexa.pdf')
plt.show()
```
